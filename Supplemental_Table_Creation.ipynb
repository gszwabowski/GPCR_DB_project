# Get Structure Information from GPCRdb

```python
df = pd.read_csv('data/364_interaction_energies_state_function_v4.txt', sep='\t')
df.fillna('None', inplace=True)
```

```python
df
```

```python
# drop entries with bad ligand placement
PDBs_toremove = [
    '4LDL',
    '4LDO',
    '7VDL',
    '7VDM',
    '7VUY',
    '7VUZ',
    '7VV0',
    '7VV3',
    '7VV4',
    '7VV5',
    '7NA7',
    '7FIG',
    '7FIH',
    '7P02',
    '7P00',
    '4U15',
    '2YCX',
    '6YVR',
    '5T04',
    '4XEE',
    '4XES',
    '4GRV'
]

to_drop = []

for i in range(len(df)):
    if df.iloc[i,:]['PDBID'] in PDBs_toremove:
        to_drop.append(i)

df.drop(to_drop, inplace = True)
```

```python
PDBids = df['PDBID']
```

```python
protein_names = []
species = []
pub_dates = []
states = []
ligand_names = []
ligand_functions = []

for PDBid in PDBids:
    print(PDBid)
    url = 'https://gpcrdb.org/services/structure/' + PDBid
    response = requests.get(url)
    data = response.json()
    protein_names.append(data['protein'])
    species.append(data['species'])
    pub_dates.append(data['publication_date'])
    states.append(data['state'])
    
    ligand_data = data['ligands'][0]
    ligand_names.append(ligand_data['name'])
    ligand_functions.append(ligand_data['function'])
    
print(protein_names)
```

```python
protein_names
```

```python
# remove suffixes from protein names
for i in range(len(protein_names)):
    sep = '_'
    protein_names[i] = protein_names[i].split(sep, 1)[0].upper()
```

```python
protein_names
```

```python
# create dataframe with retrieved data
ED_df = pd.DataFrame({
    'PDBID' : PDBids,
    'Protein Name' : protein_names,
    'Species' : species,
    'Publication Date' : pub_dates,
    'Activation State' : states,
    'Ligand Name' : ligand_names,
    'Ligand Function' : ligand_functions
})
```

```python
ED_df = ED_df.sort_values('PDBID')
```

```python
ED_df.to_csv('ED_structure_info.txt')
```

# Cross-docking Structures

```python
# read in .txt file with crossdock pdbs generated using get_crossdock_pdbs.ipynb
crossdock_df = pd.read_csv('crossdock_pdbs.txt')
```

```python
ED_df = pd.read_csv('ED_structure_info.txt', index_col = 0)
ED_df['PDBID'] = ED_df['PDBID'].astype(str)
```

```python
crossdock_df = crossdock_df.rename({'PDB' : 'PDBID'}, axis = 1)
crossdock_df['PDBID'] = crossdock_df['PDBID'].astype(str)
```

```python
crossdock_df
```

```python
ED_df
```

```python
ED_df_merged = ED_df.merge(crossdock_df, on = 'PDBID', how = 'left')
```

```python
ED_df_merged.drop(['Index', 'UniProt', 'State', 'Function', 'dock_mol'], axis = 1, inplace = True)
```

```python
ED_df_merged = ED_df_merged.rename({'crossdock_PDB' : 'Cross-docking PDBID'}, axis = 1)
```

```python
ED_df_merged = ED_df_merged.sort_values(['Protein Name', 'Ligand Name'])
```

```python
# write to word table 
import docx

# open an existing document
try:
    doc = docx.Document('ED_structure_table.docx')
except:
    doc = docx.Document()
    doc.save('ED_structure_table.docx')
    print("Previous file was corrupted or didn't exist - new file was created.")

# add a table to the end and create a reference variable
# extra row is so we can add the header row
t = doc.add_table(ED_df_merged.shape[0]+1, ED_df_merged.shape[1])

# add the header rows.
for j in range(ED_df_merged.shape[-1]):
    t.cell(0,j).text = ED_df_merged.columns[j]

# add the rest of the data frame
for i in range(ED_df_merged.shape[0]):
    for j in range(ED_df_merged.shape[-1]):
        t.cell(i+1,j).text = str(ED_df_merged.values[i,j])

# save the doc
doc.save('./ED_structure_table.docx')
```

```python
selfdock_ligands_df_merged = selfdock_ligands_df_merged[['Structure', 'Ligand Name', 'Ligand Function', 'PDBID', 'Protein Name']]
```

```python
selfdock_ligands_df_merged = selfdock_ligands_df_merged.rename({'Protein Name' : 'Target'}, axis = 1)
```

```python
# read in .txt file with crossdock pdbs generated using get_crossdock_pdbs.ipynb
crossdock_df = pd.read_csv('crossdock_pdbs.txt')
```

```python
crossdock_df = crossdock_df.rename({'PDB' : 'PDBID'}, axis = 1)
crossdock_df['PDBID'] = crossdock_df['PDBID'].astype(str)
```

```python
selfdock_ligands_df_merged = selfdock_ligands_df_merged.merge(crossdock_df, on = 'PDBID', how = 'left')
```

```python
selfdock_ligands_df_merged.drop(['Function', 'Index', 'UniProt', 'State', 'dock_mol'], axis = 1, inplace = True)
```

```python
selfdock_ligands_df_merged.loc[selfdock_ligands_df_merged['Ligand Name'] == 'Carazolol']
```

```python
structures = []
ligand_names = []
functions = []
targets = []
selfdock_PDBIDs = []
crossdock_PDBIDs = []

# get unique ligand names
for i in range(len(selfdock_ligands_df_merged)):
    ligand = selfdock_ligands_df_merged.iloc[i, 1]
    if ligand not in ligand_names:
        ligand_names.append(ligand)
        
for name in ligand_names:
    ligand_df = selfdock_ligands_df_merged.loc[selfdock_ligands_df_merged['Ligand Name'] == name]
    ligand_df.fillna('NA', inplace = True)
    #ligand_df = ligand_df.drop_duplicates(subset = ['Target'])
    display(ligand_df)
    # append structure to list of structures
    structures.append(ligand_df['Structure'].tolist()[0])
    
    # append functions
    functions.append(', '.join(ligand_df['Ligand Function'].tolist()))
    
    # append targets
    targets.append(', '.join(ligand_df['Target'].tolist()))
    
    # append self-dock PDBs
    selfdock_PDBIDs.append(', '.join(ligand_df['PDBID'].tolist()))
    
    # append cross-dock PDBS
    ligand_CD_PDBIDs = ligand_df['crossdock_PDB'].tolist()
    if 'NA' in ligand_CD_PDBIDs:
        ligand_CD_PDBIDs.remove('NA')
    crossdock_PDBIDs.append(', '.join(ligand_CD_PDBIDs))
    
```

```python
refined_ligand_df = pd.DataFrame({'Structure' : structures, 
              'Ligand Name' : ligand_names,
              'Target(s)' : targets,
              'Function(s)': functions,
              'Self-dock PDBID(s)' : selfdock_PDBIDs,
              'Cross-dock PDBID(s)' : crossdock_PDBIDs
             })

refined_ligand_df.sort_values(by = 'Target(s)', inplace = True)

refined_ligand_df.to_csv('all_selfdock_crossdock_ligands.txt', sep = '\t')
```
