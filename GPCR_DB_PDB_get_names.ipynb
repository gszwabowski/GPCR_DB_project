```python
import requests
import pandas as pd

# disable chained assignments
pd.options.mode.chained_assignment = None 
```

```python
df = pd.read_csv('data/364_interaction_energies_state_function_v3.txt', sep='\t')
df.fillna('None', inplace=True)
```

```python
display(df)
```

```python
state_df = df.iloc[:,[1,2,3]]
```

```python
state_df
```

```python
PDBIDs = list(state_df['PDBID'])

protein_names = []

for PDBid in PDBIDs:
    print(PDBid)
    url = 'https://gpcrdb.org/services/structure/' + PDBid
    response = requests.get(url)
    data = response.json()
    protein_names.append(data['protein'])
    
print(protein_names)
```

```python
import pickle

with open('protein_names.pkl', "wb") as f:
    pickle.dump(protein_names, f)
    f.close()
```
