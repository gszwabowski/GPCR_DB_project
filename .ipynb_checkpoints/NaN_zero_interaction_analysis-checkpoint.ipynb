```python
import pandas as pd
import numpy as np
```

## Data Transformation

```python
df = pd.read_csv('data/364_interaction_energies_state_function_v3.txt', sep='\t')
```

```python
# create dataframe describing df that includes counts per column
df_desc = df.describe()
```

```python
# drop non intenergysum columns from df_desc and df
for col in df_desc.columns:
    if 'sum' not in col:
        df_desc.drop([col], axis = 1, inplace = True)
        
for col in df.columns:
    if 'sum' not in col:
        df.drop([col], axis = 1, inplace = True)
```

```python
display(df_desc)
```

```python
display(df)
```

```python
df.isna() or df == 0
```

```python
# create dfs with t/f for nan and zeroes
nan_df = df.isna()
zero_df = df == 0
not_nan_zero_df = df.notna() & df != 0
```

```python
# get residue numbers
resnums = []
for col in df_desc.columns:
    resnums.append(col[:4])
```

```python
# get counts for respective bar plots
non_na_counts = df_desc.iloc[0, :]
non_na_zero_counts = not_nan_zero_df.sum()
na_counts = nan_df.sum()
zero_counts = zero_df.sum()
```

```python
list(non_na_zero_counts)
```

```python
matplotlib.rcParams['figure.figsize'] = [60, 5] #width, height (both in inches)
matplotlib.rcParams['figure.dpi'] = 300 #width, height (both in inches)

# Non-NaN count plot
fig, ax = plt.subplots()
width = 0.35  # the width of the bars

# create rectangles for bar plot
rects = ax.bar(resnums, non_na_counts, width, label=resnums, color = 'steelblue', edgecolor = "black", linewidth = 0.75)

# Add label text, rotate x-ticks by 90 degrees, set y-axis limits to (0,100)
plt.title("Non-NaN Counts per Residue Position Across all 364 Studied Structures", fontsize = 24)
ax.set_ylabel('Count of Structures', fontsize = 20)
ax.set_xlabel('Ballesteros-Weinstein Residue Number', fontsize = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)

# stop leftmost bar from crowding x axis
plt.xlim([-1,len(resnums)])
plt.show()
fig.tight_layout()
fig.savefig('plots/non_nan_counts_per_residue.jpg', bbox_inches='tight')

# NaN count plot
fig, ax = plt.subplots()
width = 0.35  # the width of the bars

# create rectangles for bar plot
rects = ax.bar(resnums, na_counts, width, label=resnums, color = 'steelblue', edgecolor = "black", linewidth = 0.75)

# Add label text, rotate x-ticks by 90 degrees, set y-axis limits to (0,100)
plt.title("NaN Counts per Residue Position Across all 364 Studied Structures", fontsize = 24)
ax.set_ylabel('Count of Structures', fontsize = 20)
ax.set_xlabel('Ballesteros-Weinstein Residue Number', fontsize = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)

# stop leftmost bar from crowding x axis
plt.xlim([-1,len(resnums)])
plt.show()
fig.tight_layout()
fig.savefig('plots/nan_counts_per_residue.jpg', bbox_inches='tight')

# Zero Count plot
fig, ax = plt.subplots()
width = 0.35  # the width of the bars

# create rectangles for bar plot
rects = ax.bar(resnums, zero_counts, width, label=resnums, color = 'steelblue', edgecolor = "black", linewidth = 0.75)

# Add label text, rotate x-ticks by 90 degrees, set y-axis limits to (0,100)
plt.title("Zero Counts per Residue Position Across all 364 Studied Structures", fontsize = 24)
ax.set_ylabel('Count of Structures', fontsize = 20)
ax.set_xlabel('Ballesteros-Weinstein Residue Number', fontsize = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)

# stop leftmost bar from crowding x axis
plt.xlim([-1,len(resnums)])
plt.show()
fig.tight_layout()
fig.savefig('plots/zero_counts_per_residue.jpg', bbox_inches='tight')

# Interaction Count plot (where residues are not NaN nor 0)
fig, ax = plt.subplots()
width = 0.35  # the width of the bars

# create rectangles for bar plot
rects = ax.bar(resnums, non_na_zero_counts, width, label=resnums, color = 'steelblue', edgecolor = "black", linewidth = 0.75)

# Add label text, rotate x-ticks by 90 degrees, set y-axis limits to (0,100)
plt.title("Non-NaN/Nonzero Counts per Residue Position Across all 364 Studied Structures", fontsize = 24)
ax.set_ylabel('Count of Structures', fontsize = 20)
ax.set_xlabel('Ballesteros-Weinstein Residue Number', fontsize = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)

# stop leftmost bar from crowding x axis
plt.xlim([-1,len(resnums)])
plt.ylim([0, 364])
plt.show()
fig.tight_layout()
fig.savefig('plots/non_nan_nonzero_counts_per_residue.jpg', bbox_inches='tight')
```

 ### removing residue positions with > 20% missing (NaN) or no (0.0) interactions

```python
nan_df = df.copy()
nan_zero_df = df.copy()

# drop columns from nan_df in which > 20% of entries are NaN
for resnum in resnums:
    intenergysum_col = resnum + '_intenergysum'
    
    if nan_df[intenergysum_col].isna().sum() > (0.20 * len(nan_df)):
        nan_df.drop([intenergysum_col], axis = 1, inplace = True)
        
    if (nan_zero_df[intenergysum_col].notna() & nan_zero_df[intenergysum_col] != 0).sum() < (0.80 * len(nan_zero_df)):
        nan_zero_df.drop([intenergysum_col], axis = 1, inplace = True)
```

```python
print('interactions per remaining residue in nan_df')
print('----------------------------------\n')
print('residue\tnon-NaN\tnon-Nan/nonzero')

residues = []
non_nans = []
non_nan_nonzeros = []

for col in nan_df.columns:
    residues.append(col[:4])
    non_nans.append(df[col].value_counts().sum())
    non_nan_nonzeros.append(df[col][df[col] != 0.0].value_counts().sum())
    print(col[:4], '\t', df[col].value_counts().sum(),'\t', df[col][df[col] != 0.0].value_counts().sum())
```

```python
int_count_df = pd.DataFrame({'residue' : residues, 'non-NaN_interactions' : non_nans, 'non-NaN_nonzero_interactions' : non_nan_nonzeros})
```
