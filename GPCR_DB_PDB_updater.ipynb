```python
import requests
import pandas as pd

# disable chained assignments
pd.options.mode.chained_assignment = None 
```

```python
df = pd.read_csv('364_interaction_energies_state_function.txt', sep='\t')
df.fillna('None', inplace=True)
```

```python
display(df)
```

```python
state_df = df.iloc[:,[1,2,3]]
```

```python
PDBIDs = list(state_df['PDBID'])
current_states = []
current_fxns = []

for PDBid in PDBIDs:
    print(PDBid)
    url = 'https://gpcrdb.org/services/structure/' + PDBid
    response = requests.get(url)
    data = response.json()
    current_states.append(data['state'])
    # print(data['ligands'][0])
    lig_dict = data['ligands'][0]
    current_fxns.append(lig_dict['function'])
```

```python
len(PDBIDs)
```

```python
state_df['State.new'] = pd.Series(current_states)
state_df['Function.new'] = pd.Series(current_fxns)
```

```python
state_df
```

```python
for i in range(len(state_df)):
    #print(state_df.iloc[0, 1], state_df.iloc[0, 3])
    if state_df.iloc[i, 1] != state_df.iloc[i, 3]:
        print('State has changed for PDBID', state_df.iloc[0,0])
        print('\tOld:', state_df.iloc[i, 1], 'New:', state_df.iloc[i, 3])
    if state_df.iloc[i, 2] != state_df.iloc[i, 4]:
        print('Function has changed for PDBID', state_df.iloc[i,0])
        print('\tOld:', state_df.iloc[i, 2], 'New:', state_df.iloc[i, 4])
```
